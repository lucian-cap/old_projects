This next section demonstrates some features of numpy.

For example, multiplying matrices $A = \begin{bmatrix} 1 & 2\\ 3 & 4\end{bmatrix}$, $B = \begin{bmatrix} -1 & 4\\ -1 & 2\end{bmatrix}$
results in $C = A\times B = \begin{bmatrix} -3 & 8\\ -7 & 20\end{bmatrix}$




```
import numpy as np

simple_list =[1, 2, 3]

A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[-1, 4], [-1, 2]])

# element wise multiplication, NOT matrix multiplication 
C_ = A * B

C = np.matmul(A, B)

print(C)
print(type(C))
print(type(simple_list))
```

Ways to create vectors, matrices using numpy

```
# row vector 
rv = np.array([1, 2, 3, 4])[np.newaxis]
print(rv)

# column vector
cv = rv.T
print(cv)


# inner product v. outer product
ip = np.matmul(rv, cv)
print('Below: inner product')
print(ip)

print('Below: outer product')
op = np.matmul(cv, rv)
print(op)
```

If $M$ is a matrix, then $M\times M^{-1} = I$

```

M = np.random.rand(4, 4)

print(M)
print('The inverse of M is: ')

invM = np.linalg.inv(M)
print(invM)

# the matrix product of M and M inverse
prod = np.matmul(M, invM)
print('M * Minv should be identity')
print(prod)
```


\begin{cases} 2x_1 + 3x_2 = 1 \\ .5x_1 - x_2 = -2 \end{cases}

$A = \begin{bmatrix} 2 & 3\\.5 & -1\end{bmatrix}_{2x2}$
$b = \begin{bmatrix} 1 \\ -2\end{bmatrix}$

in matrix notation:

Ax = b

where $x = \begin{bmatrix} x_1 \\ x_2\end{bmatrix}$

we can reparameterize

\begin{cases} y = \frac{1}{3} - \frac{2}{3}x \\  y = 2  + .5x\end{cases}


```
import matplotlib.pyplot as plt

f1 = lambda x: .3 - (2/3) * x
f2 = lambda x: 2 + .5 * x

domain = np.linspace(-2, 2, 100)
val_f1 = f1(domain)
val_f2 = f2(domain)


fig, ax = plt.subplots(1, 1)
ax.plot(domain, val_f1, '-r')
ax.plot(domain, val_f2, '-b')
ax.set_xlabel('Domain')
ax.set_ylabel('Values')
ax.set_title('System of equations')
ax.legend(('f1', 'f2'))

```
