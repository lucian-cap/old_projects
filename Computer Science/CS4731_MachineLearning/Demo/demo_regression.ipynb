```
import numpy as np
import matplotlib.pyplot as plt


X = np.array([1.7, 1.5, 2.8, 5, 1.3, 2.2, 1.3])
Y = np.array([368, 340, 665, 954, 331, 556, 376])

# visualize

fig, ax = plt.subplots(1, 1)
ax.plot(X, Y, 'rx')
ax.set_title('Our data')
ax.set_xlabel('Advertising Cost')
ax.set_ylabel('Sales')
```

```
y = lambda b0, b1, x: b0 + b1*x

b0 = 200
b1 = 100

domain = np.linspace(1, 5, 10)
values = y(b0, b1, domain)
b0 = 300
b1 = 50.2
values1 = y(b0, b1, domain)


fig, ax = plt.subplots(1, 1)
ax.plot(X, Y, 'rx')

ax.plot(domain, values, 'ro-')
ax.plot(domain, values1, 'ro-')

ax.set_title('Our data')
ax.set_xlabel('Advertising Cost')
ax.set_ylabel('Sales')
```

```
loss = lambda b0, b1: np.sum(np.abs(y(b0, b1, X) - Y))

# grid search


chb0 = np.linspace(50, 300, 100)
chb1 = np.linspace(-100, 300, 100)


v1, v2 = np.meshgrid(chb0, chb1)

all_losses = np.zeros((100, 100))

for ix in range(np.size(v1)):
  L = loss(v2.ravel()[ix], v1.ravel()[ix])
  all_losses.ravel()[ix] = L
  
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

result = np.where(all_losses == np.amin(all_losses))

plt.imshow(all_losses,  cmap=plt.cm.jet, extent=[-100,300, 50,300])
plt.clim([200, 500])

plt.scatter(chb1[result[1]],chb0[result[1]], 100)

ax.set_title('Error surface')
ax.set_xlabel('b0')
ax.set_ylabel('b1')

print(chb1[result[0]])
print(chb0[result[1]])
```

```
import scipy.optimize


loss = lambda b: np.sum(np.abs(y(b[0], b[1], X) - Y))
res = scipy.optimize.fmin(func=loss, x0=[(2, 2)])

#loss(2, 2)

b0 = res[0]
b1 = res[1]

print('best b0: ', b0)
print('best b1: ', b1)


fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.plot(X, Y, 'rx')
domain = np.linspace(1, 5, 10)
values = y(b0, b1, domain)
plt.plot(domain, values, '-bx')
```
