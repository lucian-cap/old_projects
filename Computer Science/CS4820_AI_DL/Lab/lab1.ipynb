#### 1.6. Array broadcasting

```python
time.strftime('%X %x %Z')

m1 = np.array([[1, 2, 3]])
m2 = np.array([[ 5, 1 ,3], 
               [ 1, 1 ,1], 
               [ 1, 2 ,1]])


# uncomment and complete the next line, which concatenates m1 to m2 on the 1st axis (i.e. row)
( np.concatenate((m2, m1), 0)  == np.array([[5, 1, 3], [1, 1, 1], [1, 2, 1], [1, 2, 3]])).all()
```

Let's take another look at b:

```python
b=np.array([[8, 5, 6, 1],
       [4, 3, 0, 7],
       [1, 3, 2, 9]])
```

`b` has the shape `(3, 4)`. If we want to plot three lines with four points each, we need to swap the rows with the columns.

```python
time.strftime('%X %x %Z')

b = b.transpose()

plt.plot(b)
```

### 3. Scikit-Learn

```python
from sklearn.datasets import make_circles

# Generate data of two circles and plot them out

X, y = make_circles(n_samples=1000,
                    noise=0.1,       # Standard deviation of Gaussian noise added to the data.
                    factor=0.2,      # Scale factor between inner and outer circle.
                    random_state=0)  # Seed for random number generation
```

```python
plt.figure(figsize=(5, 5))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses");
```
